#Exercise 4: Exploring Data Types on the Iris Dataset

```python
#load dataset
from sklearn.datasets import load_iris
import pandas as pd

# load dataset
iris = load_iris()

# switch to DataFrame
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the species column
df_iris['species'] = iris.target

df_iris.head()
```

```python
df_iris.info()
```

In Iris Dataset, we have 4 variables quantitatives and 1 variable qualitative:
- sepal length (cm) quantitative   CONTINUOUSÂ (Measurable)
- sepal width (cm)  quantitative   CONTINUOUSÂ (Measurable)
- petal length (cm) quantitative   CONTINUOUSÂ (Measurable)
- petal width (cm)  quantitative   CONTINUOUSÂ (Measurable)
- species           qualitative    NOMINALÂ (Named)  --> species of the flower

#Exercise 5: Basic Data Analysis - Choosen column [sepal length (cm)]

```python
column = 'sepal length (cm)'

mean_val = df_iris[column].mean()
median_val = df_iris[column].median()
mode_val = df_iris[column].mode()

#Basic Decriptive Statistics
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)

```

```python
import matplotlib.pyplot as plt

plt.hist(df_iris[column], bins=20)
plt.xlabel(column)
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

Findings:

* The mean and median of sepal length are close, indicating a fairly symmetric distribution.

* The histogram shows that most sepal lengths are concentrated between 4,5 and 7 cm.

* The dataset appears balanced and does not contain extreme outliers.

#Exercise 6: Basic Observation Skills in Data Analysis

1. Year combined with average hours per day sleeping
Analysis type: trend analysis
Allows you to see how average hours of sleep change over time:
Are Americans sleeping more or less than before? Is there a clear trend?


```python
# We verify the column names
df.columns
```

```python
#Line graph: Average hours of sleep as the years progressed

import matplotlib.pyplot as plt

# group by year and calculate the average
sleep_per_year = df.groupby('Year')['Avg hrs per day sleeping'].mean().reset_index()

plt.figure(figsize=(8,5))
plt.plot(
    sleep_per_year['Year'],
    sleep_per_year['Avg hrs per day sleeping']
)

plt.xlabel('Year')
plt.ylabel('Avg hrs per day sleeping')
plt.title('Average Sleep Duration Over Time')
plt.show()
```

The line graph shows a clear upward trend in the average number of hours Americans sleep.

```python
plt.figure(figsize=(8,5))

# Filter for men and women and calculate average sleep duration per year
men = df[df['Sex'] == 'Men'].groupby('Year')['Avg hrs per day sleeping'].mean().reset_index()
women = df[df['Sex'] == 'Women'].groupby('Year')['Avg hrs per day sleeping'].mean().reset_index()

plt.plot(men['Year'], men['Avg hrs per day sleeping'], label='Men')
plt.plot(women['Year'], women['Avg hrs per day sleeping'], label='Women')

plt.xlabel('Year')
plt.ylabel('Avg hrs per day sleeping')
plt.title('Average Sleep Duration Over Time by Sex')
plt.legend()
plt.show()
```

Analysis by Sex

* The graph shows the average number of hours of sleep per day for men and women over time.

* Persistent difference: the gap between the sexes remains relatively constant.

* Longer sleep duration in women: women, on average, sleep more hours per day in all the years observed.

* Occasional fluctuations: slight decreases are observed in the women's data for 2004, 2007, and 2013.

* Overall stability: in both sexes, the average sleep duration remains stable, with minor variations between years.
