# (80%) Datos no estructurados = desordenados

Son ricos en información, pero carecen de una estructura rígida.

No tienen una organización fija ni un formato predefinido.

No caben bien en tablas tradicionales, y por eso se guardan en bases NoSQL o sistemas más flexibles.

- Complejidad para convertirlos en información procesable y útil.

👉 CSV guarda todos los datos como texto, usando separadores (comas) para dividir columnas, sin mantener tipos de datos específicos.

Es el preferido para datos tabulares.

Es texto plano (flat file), simple, universal, liviano, fácil de leer, compatible con casi cualquier herramienta y perfecto para representar filas y columnas.

👉 JSON se usa para los datos jerárquicos y anidados.

Permite representar estructuras (objetos dentro de objetos y listas), mantiene relaciones entre los datos y es fácil de intercambiar entre sistemas.

Ej:datos de usuarios con atributos anidados (direcciones, preferencias, etc)

**orient** permite definir la estructura de salida.
Se adapta a diferentes casos de uso.  **Como elegir?**

* Split: Diccionario que contiene entradas separadas datos, índice y columnas → cuando los datos se usarán en machine learning o scikit-learn.

- **records** : Lista de diccionarios, uno por registro, sin índice → cuando el JSON se usará **en APIs o aplicaciones web** (lista de registros, fácil de leer y estándar).

 * Index: Diccionario de diccionarios con claves como índice y valores como registros.
 * Columns: Diccionario de listas con nombres de columna como claves y datos de columna como listas.

 → ambas, cuando necesitás acceso rápido por claves (útil para configuraciones o consultas específicas).


 * Values: Solo la matriz de valores → cuando solo importan los valores numéricos, sin etiquetas (máxima simpleza, mínima información extra).

**2. Data Augmentation**

Usado para aumentar datasets en un modelo de ML, añadiendo versiones ligeramente modificadas de datos ya existentes o datos sintéticos recién creados..
Ayuda a suavizar el modelo de aprendizaje automático y a reducir el sobreajuste de los datos (overfitting).

Muy común en imágenes y texto: rotar imágenes, cambiar brillo, flip horizontal/vertical.


Supongamos que tienes un dataset con imágenes de gatos, quieres entrenar un modelo de ML para que reconozca diferentes razas.
El dataset original puede contener imágenes de gatos en poses o fondos específicos, lo que limita la capacidad del modelo para generalizar a nuevas situaciones.

¿Por qué el data augmentation es clave?

1. Aumenta los datos sin salir a recolectarlos
Conseguir imágenes reales es caro y lento. Con augmentation, una foto genera muchas.

2. Reduce el overfitting
El modelo deja de memorizar imágenes específicas y aprende patrones más generales.

3. Mejora la generalización
Rotar, flipar o cambiar brillo simula el mundo real:
distintos ángulos, luces, fondos, posiciones.

4. Hace el modelo más robusto
Un gato sigue siendo gato aunque esté rotado, oscuro o parcialmente cortado.
El augmentation le enseña eso al modelo.

5. Es casi obligatorio en visión por computadora
Sin augmentation, los modelos de imágenes suelen fallar fuera del dataset de entrenamiento.

**3. Simulation-Based Generation**

Generar datos a partir de simulaciones de sistemas reales.

Ideal cuando recolectar datos reales es caro, poco práctico o imposible (Ejemplo: precios de acciones o tráfico).

Permite generar datos para escenarios específicos, probar hipótesis y analizar el comportamiento de los sistemas en diversas condiciones.


Ej: Para un proyecto de análisis financiero, necesitas datos históricos del precio de las acciones de la empresa X (datos costosos).

Solución: Generas generar datos sintéticos mediante una simulación del mercado financiero que replique los movimientos del precio de las acciones basándose en patrones históricos, tendencias del mercado y otros factores relevantes.

* Ventaja: barato, flexible, controlado, repetible.

* Desventaja: puede ser irreal, sesgado, complejo de crear, incompleto

**4. Crowdsourcing:** Recopilación de datos desde muchas personas ("multitud").

Las encuestas online masivas aportan diversidad y volumen frente a las encuestas tradicionales con una muestra limitada.


Ej: En una investigación de mercado que quiere entender las preferencias de los consumidores por un nuevo producto, decides crear una encuesta en línea para una amplia audiencia (usuarios de redes sociales, comunidades en línea y tu grupo demográfico objetivo).

Como resultado, recibes una cantidad considerable de respuestas con diversas perspectivas, comentarios y preferencias, lo que te ayuda a tomar decisiones informadas sobre las características y las estrategias de marketing de tu producto.

**Buenas prácticas**: garantizar cantidad y calidad: realismo, diversidad y calidad define el éxito del análisis o del modelo.


* Diversidad: cubrir muchos escenarios y variaciones, para evitar sesgos.

  Considere la diversidad demográfica, la diversidad geográfica y las variaciones en la distribución de datos.

* Mantener el realismo: que los datos se parezcan (imiten, reproduzcn) al mundo real.

* Ética y privacidad: no usar datos sensibles o de identificación personal sin consentimiento.

  Cumplir con las leyes y normativas de protección de datos.
  
  Tome medidias para mitigar posibles sesgos en los datos.

* Calidad: revisar y eliminar errores, outliers e inconsistencias + abordar valores faltantes mantiene la integridad del dataset y mejora la fiabilidad de cualquier análisis o modelo basado en él.

#Ejercicio: Creación de un conjunto de datos

Crear un conjunto de datos sintéticos que pueda utilizarse para un modelo de predicción de ventas minoristas.

1. Diseñar el esquema (columnas y tipos de datos)

Customer ID → alfanumérico

Customer Age → entero

Customer Sex → Male/Female

Product → string | variedad de 24 productos

Category → string | variedad de 10 categorías

Sale Amount → float

Date → fecha

Con ajustes de estacionalidad en el clima y también con feriados

```python
import numpy as np
import pandas as pd
from faker import Faker
from datetime import datetime, timedelta
from google.colab import files
import holidays

fake = Faker()
np.random.seed(42)

# -----------------------------
# Products and categories
# -----------------------------
products = [
    'Bicycle', 'Football', 'Tennis racket', 'Running shoes',
    'Basketball', 'Weights', 'Yoga mat', 'Cycling helmet',
    'Sports t-shirt', 'Running cap',
    'Wetsuit', 'Swimming goggles', 'Surfboard', 'Water ski', 'Fins',
    'Hiking backpack', 'Hiking boots', 'Sleeping bag', 'Tent', 'Trekking poles',
    'Alpine ski', 'Snowboard', 'Ski poles', 'Snow boots'
]

categories = [
    'Cycling', 'Football', 'Tennis', 'Running', 'Basketball',
    'Fitness', 'Yoga', 'Cycling', 'Running', 'Running',
    'Water sports', 'Water sports', 'Water sports', 'Water sports', 'Water sports',
    'Nature trips', 'Nature trips', 'Nature trips', 'Nature trips', 'Nature trips',
    'Snow sports', 'Snow sports', 'Snow sports', 'Snow sports'
]

# -----------------------------
# Israeli holidays
# -----------------------------
il_holidays = holidays.Israel(years=[datetime.today().year, datetime.today().year-1])

# -----------------------------
# Random date in last 13 months
# -----------------------------
def random_date_last_13_months():
    today = datetime.today()
    start_date = today - timedelta(days=395)  # ~13 months
    return fake.date_between(start_date=start_date, end_date=today)

# -----------------------------
# Generate dataset
# -----------------------------
data = []
num_customers = 200  # 200 distinct customers

for i in range(1, num_customers+1):
    customer_id = f"C{i:03d}"
    age = np.random.randint(13, 91)  # Age 13-90
    sex = np.random.choice(['Male', 'Female'], p=[0.5, 0.5])

    # Simulate multiple visits per customer
    num_visits = np.random.randint(5, 15)  # Each customer has 5-15 visits in 13 months

    for _ in range(num_visits):
        visit_date = random_date_last_13_months()
        num_products = np.random.randint(1, 6)  # Each visit: 1-5 products

        for _ in range(num_products):
            idx = np.random.randint(0, len(products))
            product = products[idx]
            category = categories[idx]

            base_amount = np.random.uniform(10, 1000)
            month = visit_date.month
            seasonal_multiplier = 1.0

            # Climate seasonality Israel
            if month in [6,7,8]:  # Summer
                if category in ['Running', 'Cycling', 'Water sports']:
                    seasonal_multiplier = 1.2
            if month in [12,1,2]:  # Winter
                if category == 'Snow sports':
                    seasonal_multiplier = 1.3
            if month in [3,4,5,9,10]:  # Spring & Fall
                if category == 'Nature trips':
                    seasonal_multiplier = 1.2

            # Israeli holidays multiplier
            if visit_date in il_holidays:
                seasonal_multiplier *= 1.2

            sale_amount = round(base_amount * seasonal_multiplier, 2)

            data.append([customer_id, age, sex, product, category, sale_amount, visit_date])

# -----------------------------
# Create DataFrame
# -----------------------------
sports_sales_df = pd.DataFrame(data, columns=['Customer ID', 'Customer Age', 'Customer Sex',
                                              'Product', 'Category', 'Sale Amount', 'Date'])

# -----------------------------
# Export CSV and JSON
# -----------------------------
sports_sales_df.to_csv('sports_sales_dataset_il.csv', index=False)
sports_sales_df.to_json('sports_sales_dataset_il.json', orient='records', date_format='iso')

# Download in Colab
files.download('sports_sales_dataset_il.csv')
files.download('sports_sales_dataset_il.json')

# Preview
print(sports_sales_df.head())



```
